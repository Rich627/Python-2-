```python
import requests as rq
r = rq.get('https://fchart.github.io/ML/Surveys.html')

r.encoding = 'utf-8'

with open('Surveys.html', 'wb') as f:
    f.write(r.content)
```

```python
from bs4 import BeautifulSoup as bs

with open('Surveys.html', 'r', encoding='utf8') as f:
    soup = bs(f, 'lxml')
    print(soup.prettify())
```
